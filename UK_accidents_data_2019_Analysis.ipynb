**Are pedestrians more safe, or less safe, on football match days in and around Old Trafford?**

```python
#The rows with missing longitudes and latitudes filled with -9999 will be removed to reduce noise and avoid outliers 
```

```python
new_cleaned_accidents = Accidents.dropna(subset=['Latitude', 'Longitude'])
new_cleaned_accidents[['Latitude', 'Longitude']].isna().any()
```

```python
#geographic coordinates of Old Trafford is lat 53.4631N, Log 2.2913W
```

```python
#calculate Distance Between Two Points on Earth
def get_distance(lat1, lat2, lon1, lon2):
     
    # np.radians which converts from degrees to radians.
    lon1 = np.radians(lon1)
    lon2 = np.radians(lon2)
    lat1 = np.radians(lat1)
    lat2 = np.radians(lat2)
      
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
 
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Radius of earth in kilometers
    r = 6371
      
    # calculate the result
    return(c * r)
```

```python
new_cleaned_accidents['Dist_from_trafford'] = get_distance(53.4631,new_cleaned_accidents['Latitude'],-2.2913, new_cleaned_accidents['Longitude'])
```

```python
near_locations = new_cleaned_accidents[new_cleaned_accidents['Dist_from_trafford']< 10]
```

**Sunrise/Sunset Conditions**

```python
#Group the dataset into Sunrise and Sunset Categories
#Sunrise occurs between 5 and 7 depending on the season 
#Sunset occurs betweeen 7-9 depending on the weather

def Hour_of_the_day(hour):
    if hour >= 5 and hour < 7:
        return "Sunrise (5-7)"
    elif hour >= 7 and hour < 19:
        return "Daytime (7-19)"
    elif hour >= 19 and hour < 21:
        return "Sunset (19-21)"
    else:
        return "nightime  (23-5)"

#Lets apply the hour_of_the_day function to the hour column. 
Accidents['Sunset/Sunrise'] = Accidents['hour'].apply(Hour_of_the_day)
```

```python
#To plot the graph showing the accident counts at different times of the day
DayConditions = ['nightime (23-5)', 'Sunset (19-21)', 'Daytime (7-19)', 'Sunrise (5-7)']
df = Accidents.groupby('Sunset/Sunrise').size().reindex(DayConditions)

# prepare barplot
fig, ax = plt.subplots(figsize=(10, 5))
colors = ['orange', 'blue', 'purple']

# plot
df.plot(kind='barh', ax=ax, color=colors)
ax.set_title('\nAccidents according to sunrise and sunset\n', fontsize=14, fontweight='bold')
ax.set(xlabel='\nTotal Count of Accidents', ylabel='Day Conditions')



plt.savefig('Accidents according to sunrise and sunset.png')
plt.show()
```

**Support**

support =  (number of rows containing A and B) / (Total number of transactions)



```python
#formula for support in function
def support(rule, accidents):
    occurance = len(rule)
    total = len(Accidents)
    
    return round(occurance / total,3)
```

**Confidence** (a.k.a. reliability, accuracy)

Confidence = Freq.(A and B)/ Freq.(A)
					     


```python
#formula for confidence in function
def confidence(data, rule, row, val):
    suppAnB = support(rule, data)
    suppA = sum(data[row] == val) / len(data)
    
    return round((suppAnB/suppA),2)
```

**Completeness**

**Lift**

lift = supp(AnB)/ (supp(A) * supp(B))

```python
#formula for confidence in function
def lift(data, rule, rowA, rowB, valA, valB):
    suppAnB = support(rule, data)
    suppA = sum(data[rowA] == valA) / len(data)
    suppB = sum(data[rowB] == valB) / len(data)
    
    return round(suppAnB/(suppA * suppB),2)
```

```python
rowA = 'Speed_limit'
rowB = 'Accident_Severity'
valA, valB = 20, 3
lift(Accidents, rule1, rowA, rowB, valA, valB)
```

**Conviction**

conviction = 1 - supp(B)/ (1- conf(AnB))

```python
#formula for confidence in function
def conviction(data, rule, rowA, rowB, valA, valB):
    suppB = sum(data[rowB] == valB) / len(data)
    conf = confidence(data, rule, rowA, valA)
    return round(((1 - suppB) / (1 - conf)),2)
```

### Stacking models

```python
from sklearn.ensemble import StackingClassifier

def get_stacking():
    level0 = list()
    level0.append(('randomf', RandomForestClassifier()))
    level0.append(('dtree', DecisionTreeClassifier()))
    level0.append(('bayes', GaussianNB()))
    level0.append(('lr', LogisticRegression()))
    level0.append(('knn', KNeighborsClassifier()))

    level1 = LogisticRegression()

    model = StackingClassifier(estimators=level0, final_estimator=level1, cv=5)
    return model
```

```python
#modifying to add stack
def get_models_s():
    models = dict()
    models['randomf'] = RandomForestClassifier()
    models['dtree'] = DecisionTreeClassifier()
    models['bayes'] = GaussianNB()
    models['lr'] = LogisticRegression()
    models['knn'] = KNeighborsClassifier()
    models['stacked'] = get_stacking()
    
    return models
```

```python
#Grab hold of the dataset.
X, y = get_dataset()

#Grab the models to use.
models = get_models_s()    

#Run and Evaluate the models.
results, names = list(), list()
for name, model in models.items():
    scores = evaluate_model(model, X, y)
    results.append(scores)
    names.append(name)
    print('>%s %.3f (%.3f)' % (name, mean(scores), std(scores)))
```
